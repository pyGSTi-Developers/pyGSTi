```python
cor_jac_orig = build_class_jacobian('C', 1)
#refactor this to match output of new code
cor_jac_orig_refactored_dict = {}
for pair in pairs:
    cor_jac_orig_refactored_dict[pair] = np.zeros((1,3))
    for (key, val) in cor_jac_orig.items():
        if key[0] == pair:
            if key[1] ==('X', 'Y'):
                cor_jac_orig_refactored_dict[pair] += np.array([val, 0, 0])
            if key[1] ==('X', 'Z'):
                cor_jac_orig_refactored_dict[pair] += np.array([0, val, 0])
            if key[1] == ('Y', 'Z'):
                cor_jac_orig_refactored_dict[pair] += np.array([0, 0, val])

cor_jac_orig_refactored = np.concatenate([array for array in cor_jac_orig_refactored_dict.values()], axis = 0)
print(cor_jac_orig_refactored_dict)
print(cor_jac_orig_refactored)
```

```python
cor_jac_orig_2 = build_class_jacobian('C', 2)

#refactor this to match output of new code
cor_jac_orig_refactored_dict_2 = {}
seen_pairs = []
for pair in pairs_2:
    cor_jac_orig_refactored_dict_2[pair] = np.zeros((1,len(cor_pauli_idxs_2)))
    for (key, val) in cor_jac_orig_2.items():
        if key[0] == pair:
            seen_pairs.append(pair)
            #get the index in cor_pauli_idxs_2 for key[1]
            try:
                cor_pauli_idx = list(cor_pauli_idxs_2).index((stim.PauliString(key[1][0]), stim.PauliString(key[1][1])))
            except ValueError:
                
                continue
            temp_array = np.zeros((1,len(cor_pauli_idxs_2)))
            temp_array[0, cor_pauli_idx] = val
            cor_jac_orig_refactored_dict_2[pair] += temp_array
            
cor_jac_orig_refactored_2 = np.concatenate([array for array in cor_jac_orig_refactored_dict_2.values()], axis = 0)
print(cor_jac_orig_refactored_dict_2)
print(cor_jac_orig_refactored_2)
```
